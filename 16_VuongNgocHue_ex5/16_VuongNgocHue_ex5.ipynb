### Bài 1: Tạo quần thể mô phỏng

Bạn hãy tạo một mẫu gồm 10000 phần tử mô phỏng chiều cao của nam thanh niên có giá trị từ 120cm - 200cm. Bạn hãy lưu kết quả vào biến **POP**.

Tính:
a. Chiều cao trung bình (kỳ vọng) của quần thể và độ lệch chuẩn về chiều cao của quần thể.

b. Tính tỷ lệ người cao trong quần thể, biết rằng thanh niên có chiều cao từ 180 trở lên được gọi là cao.

c. Vẽ histogram về chiều cao của quần thể. Theo bạn quần thể có phân phối chuẩn hay không?

d. Thử vẽ histogram và đồ thị hàm mật độ của phân phối chuẩn sử dụng tham số loc và scale bằng với kỳ vọng và độ lệch chuẩn của quần thể.

```python
import numpy as np
import math
from scipy import stats as stats
import matplotlib.pyplot as plt
num_samples = 10000
min_height = 120
max_height = 201
POP = np.random.randint(min_height, max_height, num_samples)
print(POP)
mean_of_POP = np.mean(POP)
print("Chiều cao trung bình: ",mean_of_POP)
std_POP = np.std(POP)
print("Độ lệch chuẩn: ",std_POP)
high_POP = POP[POP > 180]
print("Tỉ lệ người cao trong quần thể: {}% ".format((len(high_POP)/len(POP))*100))
plt.hist(POP, bins=50, edgecolor='black')
plt.xlabel('Chiều cao cm')
plt.ylabel('Số lượng')
plt.title('Histogram mẫu')
plt.show()

x_axis = np.arange(min_height, max_height, 0.1)
y_axis = stats.norm.pdf(x_axis, loc=mean_of_POP, scale=std_POP)
plt.hist(POP, bins=50, density=True, edgecolor='black')
plt.plot(x_axis, y_axis)
plt.xlabel('Chiều cao (cm)')
plt.ylabel('Mật độ')
plt.title('Histogram và đồ thị hàm mật độ của phân phối chuẩn')
plt.show()
```

### Bài 4: Khảo sát phân phối của trung bình mẫu

a. Thực hiện bài 2 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

b. Thực hiện bài 3 100 lần, mỗi lần bạn tính được trung bình mẫu. Vẽ đồ thị histogram cho 100 trung bình mẫu bạn tính được.

c. Vẽ histogram của hai câu a, b trong cùng một hình để so sánh. Theo bạn hình dạng của histogram thay đổi nói lên điều gì?

Gợi ý: tạo một danh sách rỗng để chứa các trung bình mẫu tính được, sử dụng hàm append() để thêm giá trị trung bình vào danh sách sau mỗi lần tính.

```python
mean_of_M20_100 = []
for i in range(100):
    M20 = np.random.choice(POP,size = (20))
    mean_of_M20_100.append(np.mean(M20))
plt.hist(mean_of_M20_100, bins=10, edgecolor='yellow')
plt.xlabel('Chiều cao trung bình (cm)')
plt.ylabel('Số lượng')
plt.title('Histogram của 100 trung bình mẫu cỡ 20')
plt.show()

mean_of_M100_100 = []
for i in range(100):
    M100 = np.random.choice(POP,size = (100))
    mean_of_M100_100.append(np.mean(M100))
plt.hist(mean_of_M100_100, bins=10, edgecolor='yellow')
plt.xlabel('Chiều cao trung bình (cm)')
plt.ylabel('Số lượng')
plt.title('Histogram của 100 trung bình mẫu cỡ 100')
plt.show()

plt.hist(mean_of_M20_100, bins=10, edgecolor='yellow', alpha=0.5)
plt.hist(mean_of_M100_100, bins=10, edgecolor='yellow', alpha=0.5)
plt.xlabel('Chiều cao trung bình (cm)')
plt.ylabel('Số lượng')
plt.title('So sánh histogram của 100 trung bình mẫu cỡ 20 và 100')
plt.legend(['Cỡ mẫu 20', 'Cỡ mẫu 100'])
plt.show()

```

### Bài 6: Khảo sát phương sai mẫu

Bạn hãy mô phỏng 1000 lần lấy mẫu có kích cỡ 100. Vẽ histogram đại lượng phương sai của các mẫu (Phương sai bằng bình phương độ lệch chuẩn)

```python
var_M100_1000 = []
for i in range(1000):
    M100 = np.random.choice(POP,size = (100))
    var_M100 = np.var(M100)
    var_M100_1000.append(var_M100)
plt.hist(var_M100_1000, bins=10, edgecolor='yellow')
plt.xlabel('Phương sai của các mẫu')
plt.ylabel('Số lượng')
plt.title('Histogram của 1000 phương sai của mẫu cỡ 100')
plt.show()
```

### Bài 7:  Bài Tập Ước Lượng

Khảo sát cân nặng (kg) của gà khi xuất chuồng, người ta cân một số con và có kết quả như sau:

> 2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8

Giả sử cân nặng của gà là biến ngẫu nhiên có phân phối chuẩn. Hãy lập khoảng tin cậy cho cân nặng trung bình của  khi xuất chuồng với độ tin cậy là 95%. Trong 2 trường hợp:

a, Biết $\sigma = 0.3$

b, Không biết $\sigma$


```python
a = np.array([2.1, 1.8, 2.0, 2.3, 1.7, 1.5, 2.0, 2.2, 1.8])
a
```

```python
mean = np.mean(a)
std_error = 0.3 / np.sqrt(len(a))
ci = (mean - 1.96 * std_error, mean + 1.96 * std_error)
print(ci)

std_dev = np.std(a, ddof=1)
t_value = stats.t.ppf(0.975, len(a) - 1)
margin_error = t_value * std_dev / np.sqrt(len(a))
ci = (mean - margin_error, mean + margin_error)
print(ci)
```
