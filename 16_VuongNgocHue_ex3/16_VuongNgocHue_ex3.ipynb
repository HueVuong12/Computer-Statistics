## NUMPY

Bạn hãy phát sinh ngẫu nhiên mảng ndarray có tên `arr` gồm 100 phần tử số nguyên có giá trị từ 0..1000. Thực hiện các thao tác sau:

### 1. Hiển thị các phần tử thỏa yêu cầu dưới đây

- 10 phần tử đầu tiên
- 5 phần tử cuối cùng
- Các phần tử thỏa mãn điều kiện chia hết cho 5 và lớn hơn 100

````python
import numpy as np
arr = np.random.randint(0,1001, size=100)
print("mang",arr)
print("10 phan tu dau tien:",arr[:10])
print("5 phan tu cuoi cung:", arr[-5:])
result = [x for x in arr if x % 5 == 0 and x > 100]
print("chia het cho 5 va lon hon 100",result)
````

### 2. Cho biết mảng trên có bao nhiêu phần tử có giá trị nhỏ hơn 100. Hiển thị các phần tử đó

````python
result = [x for x in arr if x < 100]
print(result)
````

### 3. Thực hiện thống kê:

- Tính tổng mảng
- Tìm giá trị trung bình
- Tìm giá trị trung vị
- Tìm Q1, Q2, Q3
- Tìm phân vị 20 (percentile 20)

````python
print("tong mang:",arr.sum())
print("gia tri trung binh", arr.mean())
print("trung vi:",np.median(arr))
print('Tứ phân vị [Q1, Q2, Q3]: ', np.quantile(arr, [0.25, 0.5, 0.75]))
print('Phân vị 20:', np.quantile(arr, 0.5))
````

````python
from collections import Counter
counter = Counter(arr)
most_common = counter.most_common(1)[0][0]
print(f"phan tu xuat hien nhieu nhat trong mang la {most_common}")
````

### Câu 8: Vẽ histogram

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ histogram các số ngẫu nhiên trên.

````python
# Your code here
x = np.random.normal( 0, 1, 1000)  
mt.hist(x
        , bins=100        
        , density=True     
        , stacked=True
        , color='blue'     
        , edgecolor='white' 
        )
mt.xlabel('Value')
mt.ylabel('Frequency')
mt.title('Histogram')
mt.show()
````

### Câu 9: Vẽ đồ thị hộp

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ boxplot các số ngẫu nhiên trên.

````python
# Your code here
x = np.random.rand(10000)*(4)
mt.xlabel('Value')
mt.title("Histogram")
mt.ylabel('Frequency')
mt.boxplot(x)
mt.show()
````

### Câu 10: Kết hợp trên cùng một đồ thị

Sử dụng hàm `np.random.normal(size=10000)` để phát sinh 10000 số ngẫu nghiên thuộc phân phối chuẩn. Vẽ hình thể hiện phân phối như sau:

Hướng dẫn: thêm tham số `vert=False` trong hàm vẽ `boxplot()`

````python
# Your code here
x = np.random.normal(0, 1, size= 1000)  
mt.hist(x
        , bins=100         
        , density=True     
        , stacked=False
        , color='purple'     
        , edgecolor='white'
        )
mt.ylabel('Mật độ')
mt.title('Histogram')
mt.show()
x = np.random.rand(100) * -4
mt.boxplot(x, vert=False)
mt.title("Boxplot")
mt.show()
````
